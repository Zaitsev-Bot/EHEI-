# The Learning Journey: Building a House (Part 1)

## **1. The Goal: Regression and Classification**
### **Analogy**
Imagine constructing a house:
- **Regression**: Predicting the price of the house based on features like size, location, and number of rooms.
- **Classification**: Deciding whether the house is “luxury” or “standard” based on its features.

### **Key Notes**
- **Regression**:
  - Predicts continuous outcomes, solving minimization problems.
  - Types:
    - **Linear Regression**: Straight-line relationship.
    - **Nonlinear Regression**: Parabolas or complex curves for multivariate cases.
- **Classification**:
  - Binary (Yes/No): Predicts one of two categories.
  - Multiclass: Predicts among multiple categories (e.g., digits 0–9 in digit recognition).

---

## **2. The Blueprint: Linear and Multilayer Networks**
### **Analogy**
The blueprint dictates how features combine to produce predictions:
- **Linear Models** handle simple relationships.
- **Multilayer Perceptrons (MLPs)** tackle complex relationships by layering transformations.

### **Key Notes**
- **Linear Regression Model**:
  $$ y = mx + c $$
  \( m \): Slope, \( c \): Intercept.
- **Multiple Linear Regression**:
  $$ y = \theta_0 + \theta_1x_1 + \theta_2x_2 + \dots + \theta_nx_n $$
  \( \theta \): Weights, \( x \): Features.
- **MLPs**:
  - Add **hidden layers** and **activation functions** to learn high-dimensional patterns.
  - Activation Functions:
    - **ReLU**: Handles nonlinear transformations. (Hidden layers)
    - **Sigmoid**: Useful for binary outputs. (Output Layer)
    

---




```python
# @title
import matplotlib.pyplot as plt
import numpy as np

# Define the range of z values
z = np.linspace(-10, 10, 100)

# Define the activation functions
linear = z
sigmoid = 1 / (1 + np.exp(-z))
relu = np.maximum(0, z)

# Plot Linear Activation Function
plt.figure(figsize=(5, 4))
plt.plot(z, linear, label="g(z) = z")
plt.title("Linear Activation Function")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Plot Sigmoid Activation Function
plt.figure(figsize=(5, 4))
plt.plot(z, sigmoid, label="g(z) = 1 / (1 + e^(-z))")
plt.title("Sigmoid Activation Function")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Plot ReLU Activation Function
plt.figure(figsize=(5, 4))
plt.plot(z, relu, label="g(z) = max(0, z)")
plt.title("ReLU Activation Function")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```


## **3. Evaluating Progress: Loss and Cost Functions**
### **Analogy**
The cost function evaluates how far the nails (predictions) are from the correct positions (true labels).

### **Key Notes**
- **Loss Function**: Measures error for a single data point.
- **Cost Function**: Aggregates loss over the entire dataset.
- Examples:
  - **MSE (Mean Squared Error)**: For regression.
  - **Log Loss (Binary Cross-Entropy)**: For binary classification.
  - **Softmax Cross-Entropy**: For multiclass classification.
---

### **Loss/Cost Functions with Formulas**

1. **Mean Squared Error (MSE)**:
   \
   $$ \frac{1}{N} \sum_{i=1}^N (y_{\text{true},i} - y_{\text{pred},i})^2 $$                         
   \
   Measures the average squared difference between actual and predicted values.

2. **Mean Absolute Error (MAE)**:
   \
   $$ \frac{1}{N} \sum_{i=1}^N |y_{\text{true},i} - y_{\text{pred},i}| $$                          
   \
   Measures the average absolute difference between actual and predicted values.

3. **Huber Loss**:
   \
   $$ L = \begin{cases} 0.5(y_{\text{true}} - y_{\text{pred}})^2, & \text{if } |error| \leq \delta \\ \delta |error| - 0.5\delta^2, & \text{if } |error| > \delta \end{cases} $$
   \
   A combination of MSE and MAE; robust to outliers.

4. **Binary Cross-Entropy**:
   \
   $$ -\frac{1}{N} \sum_{i=1}^N \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right] $$            
   \
   Measures the difference between predicted probabilities and true binary labels.

5. **Multiclass Cross-Entropy**:
   \
   $$ -\frac{1}{N} \sum_{i=1}^N \sum_{j=1}^C y_{i,j} \log(p_{i,j}) $$                                
   \
   Extends binary cross-entropy for multiple classes using softmax activation.

6. **Hinge Loss**:
   \
   $$ \frac{1}{N} \sum_{i=1}^N \max(0, 1 - y_i \cdot y_{\text{pred},i}) $$                           
   \
   Penalizes predictions on the wrong side of the decision boundary, commonly used in SVMs.

---

### **Loss/Cost Function Comparison Table**

| **Loss/Cost Function**          | **Simplified Formula Description**                                  | **Description**                                                                                       | **Applications**                                    | **Advantages**                                       | **Disadvantages**                                   |
|---------------------------------|---------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------|----------------------------------------------------|----------------------------------------------------|----------------------------------------------------|
| **Mean Squared Error (MSE)**    | Average squared error                                              | Measures how far predictions deviate from true values on average.                                     | Regression tasks.                                   | Penalizes large errors more than small ones.       | Sensitive to outliers; large deviations heavily penalized. |
| **Mean Absolute Error (MAE)**   | Average absolute error                                             | Measures the absolute difference between predictions and true values.                                 | Regression tasks.                                   | Robust to outliers compared to MSE.                | Gradient is constant, making optimization less stable. |
| **Huber Loss**                  | Combines squared and absolute error                                | Uses squared error for small errors and absolute error for large errors, making it robust to outliers.| Regression tasks with robustness to outliers.      | Combines benefits of MSE and MAE.                  | Requires tuning the threshold \(\delta\).          |
| **Binary Cross-Entropy**        | Logarithmic loss for binary outcomes                               | Measures the performance of a classification model with probabilities for binary outputs.             | Binary classification tasks.                       | Suitable for probabilistic outputs.                | Poor performance with imbalanced datasets.         |
| **Multiclass Cross-Entropy**    | Logarithmic loss across multiple classes                           | Extends binary cross-entropy to handle multiple classes using softmax activation.                     | Multiclass classification tasks.                   | Handles multiclass probabilistic outputs.          | Sensitive to poor initializations or imbalanced datasets. |
| **Hinge Loss**                  | Penalizes incorrect classifications with a margin                  | Penalizes predictions that are on the wrong side of the decision boundary.                            | Classification tasks, especially for SVMs.         | Effective for linear classifiers like SVMs.        | Not suitable for probabilistic models; sensitive to noisy data. |

## **4. Tools for Improvement: Gradients and Optimization**
### **Analogy**
Gradients and optimization are tools to adjust misplaced nails:
- **Gradients**: Indicate the steepest direction for improvement.
- **Jacobian**: A vector of partial derivatives for outputs.

### **Key Notes**
- **Optimization Types**:
  - **Batch Gradient Descent**: Updates weights after processing the full dataset.
  - **Stochastic Gradient Descent (SGD)**: Updates weights after each data point.
  - **Mini-Batch Gradient Descent**: Uses subsets (mini-batches) for balance.
- **Optimizers**:
  - **Adam**: Combines Momentum and RMSProp for efficiency.
- **Second Derivative**: Ensures slope stability and avoids overshooting.


## **Types of Gradient Descent**

### **Analogy**
Imagine you're building a house. You need to nail in boards (weights) using feedback from your blueprint (cost function). How you apply feedback depends on the approach:

1. **Batch Gradient Descent**:
   - **How It Works**: You calculate all the feedback (gradients) after inspecting the entire house.
   - **Formula**:
     $$ \theta = \theta - \alpha \cdot \nabla J(\theta) $$
   - **Advantages**: Stable updates; works well for small datasets.
   - **Disadvantages**: Slow and memory-intensive for large datasets.

2. **Stochastic Gradient Descent (SGD)**:
   - **How It Works**: You adjust the nails after inspecting **one board at a time**.
   - **Formula**:
     $$ \theta = \theta - \alpha \cdot \nabla J(\theta; x_i, y_i) $$
   - **Advantages**: Fast updates; works well for large datasets.
   - **Disadvantages**: High variance in updates; may cause oscillations.

3. **Mini-Batch Gradient Descent**:
   - **How It Works**: You adjust nails after inspecting **a small group of boards** (mini-batch).
   - **Formula**:
     $$ \theta = \theta - \alpha \cdot \nabla J(\theta; x_{\text{batch}}, y_{\text{batch}}) $$
   - **Advantages**: Efficient and stable; balances speed and accuracy.
   - **Disadvantages**: Requires tuning the batch size.

4. **Momentum**:
   - **How It Works**: You use the momentum from your past adjustments to smooth out the current nailing process.
   - **Formula**:
     $$ v_t = \beta v_{t-1} + (1 - \beta) \nabla J(\theta) $$
     $$ \theta = \theta - \alpha v_t $$
   - **Advantages**: Smooth convergence; faster learning.
   - **Disadvantages**: Requires extra tuning of momentum hyperparameters.

5. **Adam Optimizer**:
   - **How It Works**: Combines momentum and RMSProp to adapt nail adjustments based on the house's conditions.
   - **Formula**:
     $$ m_t = \beta_1 m_{t-1} + (1 - \beta_1) \nabla J(\theta) $$
     $$ v_t = \beta_2 v_{t-1} + (1 - \beta_2) (\nabla J(\theta))^2 $$
     $$ \theta = \theta - \alpha \frac{m_t}{\sqrt{v_t} + \epsilon} $$
   - **Advantages**: Combines the best of momentum and RMSProp.
   - **Disadvantages**: Computationally expensive.

## **5. Rules of Adjustment: Sum, Power, Product, and Chain Rules**

## **1. Sum Rule**
### **Formula**:
$$ (f(x) + g(x))' = f'(x) + g'(x) $$

### **When to Use**:
When functions are added together.

### **Example**:
If \( h(x) = x^2 + 3x \), then \( h'(x) = 2x + 3 \).

### **Analogy**:
Two workers painting separate walls: One worker paints \( f(x) \), the other paints \( g(x) \). Their combined effort is the sum of their rates.

---

## **2. Power Rule**
### **Formula**:
$$ (x^n)' = n \cdot x^{n-1} $$

### **When to Use**:
When the function involves a term like \( x^n \).

### **Example**:
If \( f(x) = x^3 \), then \( f'(x) = 3x^2 \).

### **Analogy**:
A stack of coins: Each coin represents a layer (\( n \)). The change as you remove a coin (differentiation) reduces the height by one (\( n-1 \)), leaving a smaller stack.

---

## **3. Product Rule**
### **Formula**:
$$ (f(x) \cdot g(x))' = f'(x) \cdot g(x) + f(x) \cdot g'(x) $$

### **When to Use**:
When two functions are multiplied together.

### **Example**:
If \( h(x) = x^2 \sin(x) \), then \( h'(x) = 2x \sin(x) + x^2 \cos(x) \).

### **Analogy**:
Two people working independently: \( f(x) \) assembles parts, \( g(x) \) paints them. Each contributes separately, but the total effort combines their contributions.

---

## **4. Chain Rule**
### **Formula**:
$$ (f(g(x)))' = f'(g(x)) \cdot g'(x) $$

### **When to Use**:
When one function is inside another (e.g., \( f(g(x)) \)).

### **Example**:
If \( h(x) = (3x + 2)^2 \), then \( h'(x) = 6(3x + 2) \).

### **Analogy**:
An assembly line: A factory processes \( g(x) \) (raw materials) and sends it to \( f(g(x)) \) (finishing), where the output depends on both processes.



### **Key Notes**
These rules form the mathematical foundation for **backpropagation** and feature adjustment.

# **Softmax Analogy: The Election**

Imagine there’s an **election** in which several candidates are running, and your goal is to determine how much support each candidate has based on some initial estimates of their popularity. The **softmax function** is like a process that transforms those estimates into clear, understandable percentages (probabilities) for each candidate.

---

## **Breaking it Down**

### **1. Initial Votes: The Logits**
Each candidate has some "raw vote count" (logit), which represents their current popularity. These numbers could be unbalanced and hard to compare.

- Example: Candidate A has **10 votes**, Candidate B has **5 votes**, and Candidate C has **2 votes**.

---

### **2. Amplify the Differences: Exponentials**
The softmax process applies a magic amplifier (the exponential function) to make popular candidates even more popular and less popular ones stand out less.

\[
S_i = e^{z_i}
\]

Where:
- \( z_i \): The raw score (logit) for class \( i \).
- \( e^{z_i} \): Exponentiation to amplify the score.

- Example:
  - Candidate A: \( e^{10} = 22026.47 \)
  - Candidate B: \( e^{5} = 148.41 \)
  - Candidate C: \( e^{2} = 7.39 \)

---

### **3. Normalize the Votes: Total Support**
Add up all the amplified scores to calculate the **total support**:

\[
\text{Total Support} = \sum_{j=1}^{C} e^{z_j}
\]

Each candidate’s percentage is calculated by dividing their amplified score by the total support:

\[
P_i = \frac{e^{z_i}}{\sum_{j=1}^{C} e^{z_j}}
\]

Where:
- \( P_i \): The probability of class \( i \).
- \( C \): The total number of classes (candidates).

- Example:
  - Total Support = \( 22026.47 + 148.41 + 7.39 = 22182.27 \)
  - Candidate A: \( \frac{22026.47}{22182.27} = 99.29\% \)
  - Candidate B: \( \frac{148.41}{22182.27} = 0.67\% \)
  - Candidate C: \( \frac{7.39}{22182.27} = 0.03\% \)

---

### **4. Final Output: Probabilities**
Now you have clear percentages (probabilities) that sum to 100%. These represent how likely it is that each candidate will win based on their current support.

- Candidate A is the obvious winner with the highest percentage.

---

## **Why This Makes Sense**

Softmax ensures:
1. The total support is always distributed as probabilities (percentages) that sum to 1.
2. Small differences in popularity become clearer after amplification.
3. The most popular candidate stands out, while others are given fair consideration.

---

## **How This Relates to Neural Networks**

- Each **logit** is like a candidate, representing how strongly the model believes in a particular class.
- **Softmax** converts these logits into probabilities so that you can clearly see which class (candidate) the model predicts.
- The class with the highest probability is the "winner" (the predicted class).

---

## **Key Takeaway**

Softmax is like **turning raw vote counts into election results**:
1. Amplify differences (exponentials).
2. Normalize to total votes (probabilities).
3. The winner is the candidate (class) with the highest probability.


```python
# @title

# Title: Visualizing Loss/Cost Functions and Softmax
import matplotlib.pyplot as plt
import numpy as np

# Define range of prediction errors
errors = np.linspace(-2, 2, 100)
delta = 1  # Huber Loss threshold

# Define loss functions
mse = errors**2  # Mean Squared Error
mae = np.abs(errors)  # Mean Absolute Error
huber = np.where(np.abs(errors) <= delta, 0.5 * errors**2, delta * np.abs(errors) - 0.5 * delta**2)  # Huber Loss
hinge = np.maximum(0, 1 - errors)  # Hinge Loss

# Softmax Cross-Entropy (example for 2 classes, not plotted here)
def binary_cross_entropy(p):
    return -np.log(p)

# Softmax function
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # Subtract max for numerical stability
    return exp_z / exp_z.sum(axis=0)

# Multiclass Cross-Entropy Loss
def multiclass_cross_entropy(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred))

# Plot Mean Squared Error (MSE)
plt.figure(figsize=(6, 4))
plt.plot(errors, mse, label="MSE: Mean Squared Error")
plt.title("Mean Squared Error")
plt.xlabel("Prediction Error")
plt.ylabel("Loss")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Plot Mean Absolute Error (MAE)
plt.figure(figsize=(6, 4))
plt.plot(errors, mae, label="MAE: Mean Absolute Error")
plt.title("Mean Absolute Error")
plt.xlabel("Prediction Error")
plt.ylabel("Loss")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Plot Huber Loss
plt.figure(figsize=(6, 4))
plt.plot(errors, huber, label=f"Huber Loss (delta={delta})")
plt.title("Huber Loss")
plt.xlabel("Prediction Error")
plt.ylabel("Loss")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Plot Hinge Loss
plt.figure(figsize=(6, 4))
plt.plot(errors, hinge, label="Hinge Loss")
plt.title("Hinge Loss")
plt.xlabel("Prediction Error")
plt.ylabel("Loss")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Softmax example for binary cross-entropy
probabilities = np.linspace(0.01, 0.99, 100)
bce = binary_cross_entropy(probabilities)

plt.figure(figsize=(6, 4))
plt.plot(probabilities, bce, label="Binary Cross-Entropy Loss")
plt.title("Binary Cross-Entropy Loss")
plt.xlabel("Predicted Probability (p)")
plt.ylabel("Loss")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0.5, color="red", linewidth=0.5, linestyle="--", label="Decision Boundary (p=0.5)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Example: Softmax Probabilities for Multiclass Classification
z = np.linspace(-2, 2, 100)  # Example logits
z_matrix = np.vstack([z, z + 1, z - 1])  # Simulated logits for 3 classes
softmax_probs = softmax(z_matrix)

# Plot Softmax Probabilities
plt.figure(figsize=(8, 5))
for i, probs in enumerate(softmax_probs):
    plt.plot(z, probs, label=f"Class {i + 1} Probability")
plt.title("Softmax Activation Function")
plt.xlabel("Logits (z)")
plt.ylabel("Probability")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Example: Multiclass Cross-Entropy Loss for 3 Classes
y_true = np.array([1, 0, 0])  # One-hot encoded true label
y_pred = softmax(np.array([2.0, 1.0, 0.1]))  # Predicted probabilities using softmax

# Compute Loss
loss = multiclass_cross_entropy(y_true, y_pred)
print(f"Multiclass Cross-Entropy Loss: {loss:.4f}")
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.regularizers import L2
import matplotlib.pyplot as plt

# Load dataset (example: MNIST)
from tensorflow.keras.datasets import mnist
(X, y), (_, _) = mnist.load_data()  # Load MNIST dataset
X = X.reshape(X.shape[0], -1).astype('float32') / 255.0  # Normalize and flatten

# Define a Neural Network Model
model = Sequential([
    Input(shape=(X.shape[1],)),  # Explicit Input Layer
    Dense(25, activation='relu', kernel_regularizer=L2(0.01)),  # Layer 1
    Dropout(0.5),  # Dropout
    Dense(15, activation='relu', kernel_regularizer=L2(0.01)),  # Layer 2
    Dropout(0.5),  # Dropout
    Dense(10, activation='linear')  # Output Layer
])

# Compile the Model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the Model
history = model.fit(X, y, epochs=40, validation_split=0.2)

# Visualize Loss
def plot_loss(history):
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid()
    plt.show()

plot_loss(history)
```
