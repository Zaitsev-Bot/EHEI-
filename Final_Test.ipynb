```python

class Item:
    def __init__(self, category, code, description, price):
        self.category = category
        self.code = code
        self.description = description
        self.price = price


    def __str__(self):
        return f"{self.description} - ${self.price:.2f}"




class Shop:
    def __init__(self):
        self.inventory = [
            Item("Phone", "BPCM", "Compact", 29.99), Item("Phone", "BPSH", "Clam Shell", 49.99),
            Item("Phone", "RPSS", "Robo phone - 5inch 64GB memory", 199.99),
            Item("Phone", "RPLL", "Robo phone - 6inch 256GB memory", 499.99),
            Item("Phone", "YPLS", "Y-phone standard 6 inch 64GB memory", 549.99),
            Item("Phone", "YPLL", "Y-phone deluxe 6 inch 256GB memory", 649.99),
            Item("Tablet", "RTMS", "RoboTab - 8 inch screen 64GB memory", 149.99),
            Item("Tablet", "RTLM", "RoboTab - 10 inch screen 128GB memory", 299.99),
            Item("Tablet", "YTLM", "Y-tab standard - 10 inch screen 128GB memory", 499.99),
            Item("Tablet", "YTLL", "Y-tab deluxe - 10 inch screen 256GB memory", 599.99),
            Item("SIM Card", "SMNO", "Sim free (no SIM card)", 0.00),
            Item("SIM Card", "SMPG", "Pay as you go (with SIM card)", 9.99),
            Item("Case", "CSST", "Standard", 0.00),
            Item("Case", "CSLX", "Luxury", 50.00),
            Item("Charger", "CGCR", "Car", 19.99),
            Item("Charger", "CGHM", "Home", 15.99)
        ]


    def display_phones_and_tablets(self):
        print("\nAvailable Phones and Tablets:")
        items = [item for item in self.inventory if item.category in ["Phone", "Tablet"]]
        for index, item in enumerate(items, start=1):
            print(f"{index}. {item}")
        return items


    def choose_phone_or_tablet(self):
        items = self.display_phones_and_tablets()
        try:
            choice = int(input("\nEnter the number of the Phone or Tablet you want to choose: ")) - 1
            if 0 <= choice < len(items):
                selected_item = items[choice]
                print(f"You selected: {selected_item}")
                return selected_item
            else:
                print("Invalid selection. Please try again.")
                return None
        except ValueError:
            print("Invalid input. Please enter a number.")
            return None


    def choose_sim_option(self, chosen_device):
        if chosen_device.category != "Phone":
            return None

        print("\nSIM Options:")
        sim_options = [item for item in self.inventory if item.category == "SIM Card"]
        for index, sim in enumerate(sim_options, start=1):
            print(f"{index}. {sim}")

        try:
            choice = int(input("\nEnter the number of the SIM option you want to choose: ")) - 1
            if 0 <= choice < len(sim_options):
                selected_sim = sim_options[choice]
                print(f"You selected: {selected_sim}")
                return selected_sim
            else:
                print("Invalid selection. Please try again.")
                return None
        except ValueError:
            print("Invalid input. Please enter a number.")
            return None


    def choose_case(self):
        print("\nCase Options:")
        case_options = [item for item in self.inventory if item.category == "Case"]
        for index, case in enumerate(case_options, start=1):
            print(f"{index}. {case}")

        try:
            choice = int(input("\nEnter the number of the Case option you want to choose: ")) - 1
            if 0 <= choice < len(case_options):
                selected_case = case_options[choice]
                print(f"You selected: {selected_case}")
                return selected_case
            else:
                print("Invalid selection. Please try again.")
                return None
        except ValueError:
            print("Invalid input. Please enter a number.")
            return None


    def choose_chargers(self):
        print("\nCharger Options:\n1. Only Car Charger\n2. Only Home Charger\n3. Both Car and Home Chargers")
        while True:
            choice = input("Enter your choice (1, 2, or 3): ")
            if choice == "1":
                selected_chargers = [item for item in self.inventory if item.code == "CGCR"]
                print(f"You selected: {selected_chargers[0]}")
                return selected_chargers
            elif choice == "2":
                selected_chargers = [item for item in self.inventory if item.code == "CGHM"]
                print(f"You selected: {selected_chargers[0]}")
                return selected_chargers
            elif choice == "3":
                selected_chargers = [item for item in self.inventory if item.category == "Charger"]
                print(f"You selected: {', '.join(str(charger) for charger in selected_chargers)}")
                return selected_chargers
            print("Invalid selection. Please select 1, 2, or 3.")


    def calculate_total(self, chosen_device, chosen_sim=None, chosen_case=None, chosen_chargers=None):
        items_purchased = []
        total_price = 0.0


        if chosen_device:
            items_purchased.append(chosen_device)
            total_price += chosen_device.price


        if chosen_sim:
            items_purchased.append(chosen_sim)
            total_price += chosen_sim.price


        if chosen_case:
            items_purchased.append(chosen_case)
            total_price += chosen_case.price


        if chosen_chargers:
            items_purchased.extend(chosen_chargers)
            total_price += sum(charger.price for charger in chosen_chargers)


        print("\n--- Items Purchased ---")
        for item in items_purchased:
            print(f"{item.category} - {item.description} - ${item.price:.2f}")

        print(f"\nTotal Price: ${total_price:.2f}")
        return total_price


    def order_multiple_devices(self):
        total_price = 0.0
        discount_amount = 0.0
        all_items_purchased = []
        device_count = 0


        while True:
            print("\nOrder a Device:")
            chosen_device = self.choose_phone_or_tablet()
            if chosen_device:
                device_count += 1
                all_items_purchased.append(chosen_device)
                price = chosen_device.price


                if device_count > 1:
                    discount = price * 0.10
                    discount_amount += discount
                    price -= discount


                total_price += price


                chosen_sim = self.choose_sim_option(chosen_device)
                if chosen_sim:
                    all_items_purchased.append(chosen_sim)
                    total_price += chosen_sim.price


                chosen_case = self.choose_case()
                if chosen_case:
                    all_items_purchased.append(chosen_case)
                    total_price += chosen_case.price


                chosen_chargers = self.choose_chargers()
                if chosen_chargers:
                    all_items_purchased.extend(chosen_chargers)
                    total_price += sum(charger.price for charger in chosen_chargers)


            while True:
                another_device = input("Would you like to order another device? (yes/no): ").strip().lower()
                if another_device in ["yes", "no"]:
                    break
                else:
                    print("Invalid answer. Please respond with 'yes' or 'no'.")


            if another_device == "no":
                break


        print("\n--- Final Order Summary ---")
        for item in all_items_purchased:
            print(f"{item.category} - {item.description} - ${item.price:.2f}")
        print(f"\nTotal Price before discounts: ${total_price + discount_amount:.2f}")
        print(f"Total Discount Applied: ${discount_amount:.2f}")
        print(f"Final Price After Discount: ${total_price:.2f}")
        return total_price




def main():
    shop = Shop()
    print("Welcome to the Shop!")
    shop.order_multiple_devices()


if __name__ == "__main__":
    main()
```
